#### Problema 1
Problema 1: El número desconocidoEnunciado: La suma de un número más 9 es igual a 21. ¿Cuál es ese número?Paso 1: Análisis y ResoluciónDefinir la variable: Sea $x$ el número que buscamos.Plantear la ecuación: El enunciado dice "un número ($x$) más (+) el número 9 es igual (=) a 21".$$x + 9 = 21$$Despejar la variable: Para dejar a la $x$ sola, restamos 9 en ambos lados de la ecuación:$$x = 21 - 9$$Solución:$$x = 12$$

```python
# Problema 1: Hallar el número desconocido
# resultado_total = 21
# numero_conocido = 9
resultado_total=int(input("Ingrese el resultado total:"))
numero_conocido=int(input("Ingrese el numero conocido:"))
# Despejemos: numeroicognito = resultad0

# Despeje: x = 21 - 9
x = resultado_total - numero_conocido

print(f"El número buscado en el problema 1 es: {x}")
```

Problema 2: Las edades de Pedro y AnaE: Pedro tiene el doble de edad que su hermana Ana. Si entre los dos suman 36 años, ¿cuál es la edad de cada uno?Paso 1: Análisis y ResoluciónDefinir las variables:Edad de Ana: $a$Edad de Pedro: $2a$ (porque es el doble que la de Ana).Plantear la ecuación: La suma de ambas edades es 36.$$a + 2a = 36$$Simplificar: Sumamos los términos semejantes ($1a + 2a = 3a$).$$3a = 36$$Despejar la variable: El 3 que multiplica a $a$ pasa al otro lado dividiendo.$$a = \frac{36}{3}$$Solución:Ana: $a = 12$ años.Pedro: $2 \times 12 = 24$ años.

```python
# Problema 2: Edades de Pedro y Ana
# suma_edades = 36
suma_edades=int(input("Ingrese la suma de las edades:"))
# Despeje de la edad de Ana (a = 36 / 3)
edad_ana = suma_edades / 3
# Edad de Pedro es el doble
edad_pedro = edad_ana * 2

print(f"Ana tiene {int(edad_ana)} años.")
print(f"Pedro tiene {int(edad_pedro)} años.")
```
